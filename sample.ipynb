```python
# 导入pytorch库
import torch
import torchvision

# 1. 查看PyTorch版本
print("PyTorch版本:", torch.__version__)
# 2. 查看torchvision版本
print("TorchVision版本:", torchvision.__version__)
# 3. 检查CUDA是否可用（如果你安装的是带CUDA的版本）
print("CUDA是否可用:", torch.cuda.is_available())
# 4. 如果有GPU，查看可用GPU数量
if torch.cuda.is_available():
    print("可用GPU数量:", torch.cuda.device_count())
    print("当前使用GPU名称:", torch.cuda.get_device_name(0))
```
