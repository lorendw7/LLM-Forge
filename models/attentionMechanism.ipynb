### Implementing a compact self-attention Python class

```python
import torch.nn as nn
class SelfAttention_v1 (nn.Module):
    def __init__(self, d_int, d_out):
        super().__init__()
        self.W_query = nn.Parameter(torch.rand(d_int, d_out))
        self.W_key = nn.Parameter(torch.rand(d_int, d_out))
        self.W_value = nn.Parameter(torch.rand(d_int, d_out))

    def forward(self, x):
        keys = x @ self.W_key
        values = x @ self.W_value
        queries = x @ self.W_query
        attn_scores = queries @ keys.T
        attn_weights = torch.softmax(attn_scores / keys.shape[-1]**0.5, dim=-1)
        context_vec = attn_weights @ values
        return context_vec
```

###  Implementing a compact causal attention class

```python
batch = torch.stack((inputs, inputs), dim=0)
print(batch.shape)
```

```python
class CausalAttention(nn.Module):
    def __init__(self, d_in, d_out, context_length,
                 dropout, qkv_bias=False):
        super().__init__()
        self.d_out = d_out
        self.W_query = nn.Linear(d_in, d_out, qkv_bias)
        self.W_key = nn.Linear(d_in, d_out, qkv_bias)
        self.W_value = nn.Linear(d_in, d_out, qkv_bias)
        self.dropout = nn.Dropout(dropout)
        self.register_buffer('mask',
                             torch.tril(torch.ones(context_length, context_length),diagonal=1))

    def forward(self, x):
        b, num_tokens, d_in = x.shape
        queries = self.W_query(x)
        keys = self.W_key(x)
        values = self.W_value(x)

        attn_scores = queries @ keys.transpose(1, 2)
        attn_scores.masked_fill_(self.mask.bool()[:num_tokens, :num_tokens], -torch.inf)
        attn_weights = torch.softmax(attn_scores / keys.shape[-1] ** 0.5, dim=-1)
        attn_weights = self.dropout(attn_weights)
        context_vec = attn_weights @ values

        return context_vec
```

```python
torch.manual_seed(123)
context_length = batch.shape[1]
ca = CausalAttention(d_in, d_out, context_length, 0.0)
context_vec = ca(batch)
print("context_vec.shape:", context_vec.shape)
```

### Extending single-head attention to multi-head attention

```python
# A wrapper class to implement multi-head attention
class MultiHeadAttentionWrapper(nn.Module):
    def __init__(self, d_in, d_out, context_length,
                dropout, num_heads, qkv_bias=False):
        super().__init__()
        self.heads = nn.ModuleList([
            CausalAttention(d_in, d_out, context_length, dropout, qkv_bias)
            for _ in range(num_heads)
        ])

    def forward(self, x):
        return torch.cat([head(x) for head in self.heads], dim=-1)
```

```python
mha = MultiHeadAttentionWrapper(d_in, d_out, context_length, 0.0, num_heads=2)
context_vec = mha(batch)
print(context_vec)
print("context_vec.shape:", context_vec.shape)
```

### Implementing multi-head attention with weight splits

```python
import torch
import torch.nn as nn


class MultiHeadAttention(nn.Module):
    """
    Multi-head self-attention (decoder-style, with causal / lower-triangular mask).

    Core idea:
        Allows the model to attend to information from different positions
        in the sequence while letting each attention head focus on different
        representation subspaces.

    Args:
        d_in: Input feature dimension (number of features per token)
        d_out: Output feature dimension (must be divisible by num_heads)
        context_length: Maximum sequence length used to build the causal mask
        dropout: Dropout probability to reduce overfitting
        num_heads: Number of attention heads
        qkv_bias: Whether to use bias in the Q/K/V linear projection layers
    """
    def __init__(self, d_in, d_out,
                 context_length, dropout, num_heads, qkv_bias=False):
        super().__init__()  # Initialize the parent nn.Module

        # Check that d_out can be evenly split across all attention heads
        assert (d_out % num_heads == 0), \
            "d_out must be divisible by num_heads so each head has the same dimension"

        # Core configuration
        self.d_out = d_out                    # Total output feature dimension
        self.num_heads = num_heads            # Number of attention heads
        self.head_dim = d_out // num_heads    # Dimension of each individual head

        # Q / K / V projection layers: map input dimension d_in to output dimension d_out
        self.W_query = nn.Linear(d_in, d_out, bias=qkv_bias)  # Query projection
        self.W_key = nn.Linear(d_in, d_out, bias=qkv_bias)    # Key projection
        self.W_value = nn.Linear(d_in, d_out, bias=qkv_bias)  # Value projection

        self.out_proj = nn.Linear(d_out, d_out)  # Final projection after concatenating all heads
        self.dropout = nn.Dropout(dropout)       # Dropout applied to attention weights

        # Register the causal mask as a buffer (not updated by gradients)
        # torch.triu creates an upper-triangular matrix
        # diagonal=1 means positions above the main diagonal are 1 (future tokens),
        # while the diagonal and below are 0
        # This prevents each token from attending to future tokens
        self.register_buffer(
            "mask",
            torch.triu(torch.ones(context_length, context_length), diagonal=1)
        )

    def forward(self, x):
        """
        Forward pass for multi-head self-attention.

        Args:
            x: Input tensor of shape [batch_size, num_tokens, d_in]
               - batch_size: number of samples in the batch
               - num_tokens: sequence length (must be <= context_length)
               - d_in: input feature dimension per token

        Returns:
            context_vec: Output tensor of shape [batch_size, num_tokens, d_out]
        """
        # Extract input dimensions:
        # b = batch size
        # num_tokens = current sequence length
        # d_in = input feature dimension
        b, num_tokens, d_in = x.shape

        # Step 1: Generate Q / K / V matrices by projecting the input
        queries = self.W_query(x)  # Shape: [b, num_tokens, d_out]
        keys = self.W_key(x)       # Shape: [b, num_tokens, d_out]
        values = self.W_value(x)   # Shape: [b, num_tokens, d_out]

        # Step 2: Split Q / K / V into multiple attention heads
        # Shape change:
        # [b, num_tokens, d_out] -> [b, num_tokens, num_heads, head_dim]
        keys = keys.view(b, num_tokens, self.num_heads, self.head_dim)
        values = values.view(b, num_tokens, self.num_heads, self.head_dim)
        queries = queries.view(b, num_tokens, self.num_heads, self.head_dim)

        # Step 3: Move num_heads to the second dimension
        # Shape change:
        # [b, num_tokens, num_heads, head_dim] -> [b, num_heads, num_tokens, head_dim]
        # This makes it easier to compute attention independently for each head
        keys = keys.transpose(1, 2)
        values = values.transpose(1, 2)
        queries = queries.transpose(1, 2)

        # Step 4: Compute attention scores (Q @ K^T)
        # keys.transpose(2, 3) changes shape to:
        # [b, num_heads, head_dim, num_tokens]
        # Resulting attention scores shape:
        # [b, num_heads, num_tokens, num_tokens]
        # Each position (i, j) represents how much token i attends to token j
        attn_scores = queries @ keys.transpose(2, 3)

        # Step 5: Apply the causal mask to block future tokens
        # Slice the mask to match the current sequence length
        # True values indicate positions that should be masked
        mask_bool = self.mask.bool()[:num_tokens, :num_tokens]

        # Fill masked positions with negative infinity
        # After softmax, these positions will get probability 0
        attn_scores.masked_fill_(mask_bool, -torch.inf)

        # Step 6: Compute attention weights (scaling + softmax + dropout)
        # Scale by sqrt(head_dim) to avoid overly large values before softmax
        attn_weights = torch.softmax(attn_scores / keys.shape[-1]**0.5, dim=-1)

        # Apply dropout to attention weights for regularization
        attn_weights = self.dropout(attn_weights)

        # Step 7: Compute context vectors (attention weights @ V)
        # Shape:
        # [b, num_heads, num_tokens, num_tokens] @ [b, num_heads, num_tokens, head_dim]
        # -> [b, num_heads, num_tokens, head_dim]
        context_vec = (attn_weights @ values)

        # Move num_heads back after attention computation
        # [b, num_heads, num_tokens, head_dim] -> [b, num_tokens, num_heads, head_dim]
        context_vec = context_vec.transpose(1, 2)

        # Step 8: Concatenate all attention heads
        # contiguous() ensures memory layout is correct before view()
        # Shape:
        # [b, num_tokens, num_heads, head_dim] -> [b, num_tokens, d_out]
        context_vec = context_vec.contiguous().view(b, num_tokens, self.d_out)

        # Step 9: Final linear projection on the concatenated output
        context_vec = self.out_proj(context_vec)

        return context_vec
```
